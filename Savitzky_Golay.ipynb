```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_daily = pd.read_excel('../other/AC_DAILY.xlsx')
cols = df_daily.columns.tolist()
cols = [col.replace(' ', '_') for col in cols]
df_daily.columns = cols
```

```python
propnum = 'PAVGTJAB8R'
y_oil = df_daily[df_daily['PROPNUM'] == propnum]['OIL'].reset_index(drop=True)
x = range(len(y_oil))
```

```python
from Savitzky_Golay_Filter import savitzky_golay
yhat = savitzky_golay(y_oil.values, 31, 5)
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.plot(x,y_oil)
ax.plot(x,yhat, color='red')
```
